```python
import spacy

nlp = spacy.load("en_core_web_sm")

def is_meaningful(sentence):
    doc = nlp(sentence)


    if len(doc) < 2:
        return False


    if not any(token.pos_ in ["NOUN", "PRON"] for token in doc):
        return False


    if not any(token.pos_ in ["VERB"] for token in doc):
        return False

    if not any(token.dep_ in ["ROOT", "xcomp", "ccomp"] for token in doc):
        return False


    return True


user_input = input("Enter a sentence: ")


if is_meaningful(user_input):
    print("The sentence is meaningful.")
else:
    print("The sentence is not meaningful.")
```
