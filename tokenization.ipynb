```python
import os
import re
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize, sent_tokenize
import docx
from PyPDF2 import PdfFileReader

# Function to open files in PDF, TXT, or DOCX format
def open_file(file_path):
    file_extension = os.path.splitext(file_path)[1].lower()

    if file_extension == '.pdf':
        with open(file_path, 'rb') as file:
            pdf_reader = PdfFileReader(file)
            text = ''
            for page_num in range(pdf_reader.numPages):
                text += pdf_reader.getPage(page_num).extractText()
        return text

    elif file_extension == '.txt':
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()

    elif file_extension == '.docx':
        doc = docx.Document(file_path)
        text = ''
        for paragraph in doc.paragraphs:
            text += paragraph.text + '\n'
        return text

    else:
        return None

# Provided text for stemmer and lemmatizer program
provided_text = "they lay back on the San Francisco grass and looked at the stars and their"

file_path = 'Once on a Time by A. A. Milne.txt'
file_text = open_file(file_path)
if file_text is not None:
    print("File opened successfully.")
else:
    print("Unsupported file format.")

# Combine provided text with file text
file_text += " " + provided_text

words = word_tokenize(file_text)

# Total number of tokens
total_tokens = len(words)

# Convert to lowercase
lowercase_words = [word.lower() for word in words]

# Articles, Punctuations, URLs, Pronouns
articles = [word for word in lowercase_words if word in ['a', 'an', 'the']]
punctuations = [word for word in lowercase_words if not word.isalnum()]
urls = [word for word in lowercase_words if re.match(r'^https?://', word)]
pronouns = [word for word in lowercase_words if word in ['i', 'you', 'he', 'she', 'it', 'we', 'they', 'me', 'him', 'her', 'us', 'them']]

# Stemming
porter_stemmer = PorterStemmer()
stemmed_words = [porter_stemmer.stem(word) for word in lowercase_words]

# Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_words = [wordnet_lemmatizer.lemmatize(word) for word in lowercase_words]

# Function to generate k-shingles
def shingling(text, k):
    shingles = set()
    words = word_tokenize(text)
    for i in range(len(words) - k + 1):
        shingle = ' '.join(words[i:i+k])
        shingles.add(shingle)
    return shingles

k = 2 
shingles = shingling(file_text, k)

# Tokenize the file by sentence and print total number of sentences
sentences = sent_tokenize(file_text)
total_sentences = len(sentences)

# Print results
print("Total number of tokens:", total_tokens)
print("Articles:", articles)
print("Punctuations:", punctuations)
print("URLs:", urls)
print("Pronouns:", pronouns)
print("Stemmed words:", stemmed_words)
print("Lemmatized words:", lemmatized_words)
# print(f"{k}-shingles:", shingles)
print("Total number of sentences:", total_sentences)
```
